```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Sum
sum_result = num1 + num2
print("Sum:", sum_result)

# Difference
diff_result = num1 - num2
print("Difference:", diff_result)

# Product
prod_result = num1 * num2
print("Product:", prod_result)

# Quotient
if num2 != 0:  # Avoid division by zero
    quot_result = num1 / num2
    print("Quotient:", quot_result)
else:
    print("Quotient: Cannot divide by zero")

# Remainder
if num2 != 0:
    rem_result = num1 % num2
    print("Remainder:", rem_result)
else:
    print("Remainder: Cannot divide by zero")
```

```python
import random  # to generate random number

number = random.randint(1, 20)  # random number between 1 and 20
guess = None
attempts = 0

while guess != number:
    guess = int(input("Guess the number (1-20): "))
    attempts += 1
    if guess < number:
        print("Too low!")
    elif guess > number:
        print("Too high!")
    else:
        print(f"Correct! You guessed it in {attempts} tries.")
```
